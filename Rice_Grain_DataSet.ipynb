Labelling the target column with numbers

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Load your dataset with 3000 records into a DataFrame, assuming it's in a file named 'your_dataset.csv'
df = pd.read_csv('rice_data.csv')

# Initialize and fit the LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(df['Class'])

# Transform the 'Class' column and store the encoded values back into the DataFrame
df['encoded_class'] = label_encoder.transform(df['Class'])

# Display the DataFrame with encoded values
print(df)
```

```python
df1 = df.drop(['Class'], axis=1)
df1
```

# Data Visualizations

```python
feature_df = df1[['Area', 'Perimeter', 'Major_Axis_Length', 'Minor_Axis_Length',
       'Eccentricity', 'Convex_Area', 'Extent']]
X = np.asarray(feature_df)
X
```

```python
y = np.asarray(df1['encoded_class'])
y
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train , X_test ,y_train , y_test = train_test_split(X , y , test_size=0.2, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Training a model (SVM with Scikit learn)

```python
from sklearn import svm

classifier = svm.SVC(kernel='linear' , gamma = 'auto' , C = 2)
classifier.fit(X_train, y_train)

y_predict = classifier.predict(X_test)
```

# Evaluating the results

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predict))
```
